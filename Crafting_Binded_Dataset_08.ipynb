```python
import pandas as pd
```

```python
epl_data_featured = pd.read_csv('EPL_Data_featured.csv')

epl_odds_data = pd.read_csv('Odds_data/EPL_Odds_data.csv')
```

```python
# Concatenating the DataFrames by columns
df = pd.concat([epl_odds_data, epl_data_featured], axis=1)
```

```python
df.info()
```

```python
# Convert 'Home Team' to lowercase
df['Home Team'] = df['Home Team'].str.lower()

# Convert 'home_team' to lowercase
df['home_team'] = df['home_team'].str.lower()

# Find matching rows based on the lowercase values
matches = df['Home Team'] == df['home_team']

# Filter to get only the rows where 'Home Team' matches 'home_team'
aligned_rows = df[matches]
```

```python
aligned_rows
```

```python
# Find non-matching rows based on the lowercase values
misaligned = df['Home Team'] != df['home_team']

# Filter to get only the rows where 'Home Team' does not match 'home_team'
misaligned_rows = df[misaligned]
```

```python
misaligned_rows = df.loc[misaligned, ['Match Date', 'Home Team', 'home_team']]
```

```python
#checking 'misalignment'
misaligned_rows
```

```python
df.columns
```

```python
columns_to_drop = ['League Division', 'Match Date', 'Home Team', 'Away Team']
```

```python
# Drop the specified columns
df_final = df.drop(columns=columns_to_drop, axis=1)
```

```python
import pandas as pd

# Assuming you have a DataFrame named df

# Desired column order
desired_order = ['week', 'date', 'home_team', 'home_xg', 'score', 'away_xg', 
                 'away_team', 'referee', 'game_id', 'home_team_elo', 
                 'away_team_elo', 'season', 'home_starters', 'away_starters']

# Append any additional columns that are not listed in the desired order
# This ensures that all columns are retained in the DataFrame
additional_columns = [col for col in df_final.columns if col not in desired_order]
new_column_order = desired_order + additional_columns

# Reorder the DataFrame
df_clean = df_final[new_column_order]

# Now df has the columns in the desired order
```
