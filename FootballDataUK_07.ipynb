```python

import pandas as pd
import time
import requests
import pandas as pd
import numpy as np

import warnings

from datetime import datetime, timedelta


warnings.filterwarnings('ignore')
```

```python
file_path = 'Odds_data/E0_1718.csv'
df1718 = pd.read_csv(file_path)

file_path = 'Odds_data/E0_1819.csv'
df1819 = pd.read_csv(file_path)

file_path = 'Odds_data/E0_1920.csv'
df1920 = pd.read_csv(file_path)

file_path = 'Odds_data/E0_2021.csv'
df2021 = pd.read_csv(file_path)

file_path = 'Odds_data/E0_2122.csv'
df2122 = pd.read_csv(file_path)

file_path = 'Odds_data/E0_2223.csv'
df2223 = pd.read_csv(file_path)
```

```python
# Since the user wants to retain all columns and records, we need to set ignore_index to True.
# This will ensure that the concatenated dataframe has a new integer index.

# Concatenating the dataframes along rows with a new index
concatenated_dfs = pd.concat([df1718, df1819, df1920, df2021, df2122, df2223], ignore_index=True)

concatenated_dfs
```

```python
concatenated_dfs.info()
```

```python
# Columns to be dropped as provided by the user
columns_to_drop = [
    'BbOU', 'BbMx>2.5', 'BbAv>2.5', 'BbMx<2.5', 'BbAv<2.5',
    'GB>2.5', 'GB<2.5', 'B365>2.5', 'B365<2.5', 'P>2.5',
    'P<2.5', 'Max>2.5', 'Max<2.5', 'Avg>2.5', 'Avg<2.5',
    'FTHG', 'HG', 'FTAG', 'AG', 'FTR', 'Res', 'HTHG', 'HTAG', 'HTR',
    'Attendance', 'Referee', 'HS', 'AS', 'HST', 'AST', 'HHW', 'AHW',
    'HC', 'AC', 'HF', 'AF', 'HFKC', 'AFKC', 'HO', 'AO', 'HY', 'AY', 'HR', 'AR'
] 


# Dropping the specified columns from the concatenated dataframe

for col in columns_to_drop:
    concatenated_dfs[col] = 0

# Dropping the columns
cleaned_df = concatenated_dfs.drop(columns=columns_to_drop)

cleaned_df.info()  # Display the head of the cleaned dataframe to verify.
```

```python


# Now, dropping columns with NaN values
final_df = cleaned_df.dropna(axis=1, how='any')

print(final_df.columns)  # Display the head of the dataframe to verify.
```

```python
# Mapping of abbreviated column names to full names as provided by the key
column_name_mapping = {
    'Div': 'League Division',
    'Date': 'Match Date',
    'HomeTeam': 'Home Team',
    'AwayTeam': 'Away Team',
    'B365H': 'Bet365 Home Win Odds',
    'B365D': 'Bet365 Draw Odds',
    'B365A': 'Bet365 Away Win Odds',
    'BWH': 'Bet&Win Home Win Odds',
    'BWD': 'Bet&Win Draw Odds',
    'BWA': 'Bet&Win Away Win Odds',
    'IWH': 'Interwetten Home Win Odds',
    'IWD': 'Interwetten Draw Odds',
    'IWA': 'Interwetten Away Win Odds',
    'PSH': 'Pinnacle Home Win Odds',
    'PSD': 'Pinnacle Draw Odds',
    'PSA': 'Pinnacle Away Win Odds',
    'WHH': 'William Hill Home Win Odds',
    'WHD': 'William Hill Draw Odds',
    'WHA': 'William Hill Away Win Odds',
    'VCH': 'VC Bet Home Win Odds',
    'VCD': 'VC Bet Draw Odds',
    'VCA': 'VC Bet Away Win Odds',
    'PSCH': 'Pinnacle Closing Home Win Odds',
    'PSCD': 'Pinnacle Closing Draw Odds',
    'PSCA': 'Pinnacle Closing Away Win Odds'
}

final_df = final_df.rename(columns=column_name_mapping)
```
